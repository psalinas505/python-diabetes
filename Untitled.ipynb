```python
#import libraries
from IPython.display import Image
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#import data sets
female_diabetes = pd.read_csv('https://raw.githubusercontent.com/psalinas505/python-diabetes/main/diabetes.csv')
all_diabetes = pd.read_csv('https://raw.githubusercontent.com/psalinas505/python-diabetes/main/diabetic_data.csv')
#import panel - this is what allows us to create the GUI
import panel as pn
#run extension - this needs to be done in every block where the panel will be used
pn.extension()
pn.extension()
#setting up the panel by adding title and widgets such as selection fields
#column can display options; this column displays the title
column = pn.Column('# Data Comparison Tool')
#these widgets allow for selection of graph axis and color of scatter points
w1 = pn.widgets.Select(name='Select 1st field to compare', options=female_diabetes.columns.tolist())
w2 = pn.widgets.Select(name='Select 2nd field to compare', options=female_diabetes.columns.tolist())
color = pn.widgets.ColorPicker(name='Select Color', value='#000000')
def show_plot(x_value, y_value, color_value):
    
    fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    message.value = "show_plot"
    plt.scatter(female_diabetes[w1.value], female_diabetes[w2.value], c=color.value, s=100)
    plt.xlabel(w1.value)
    plt.ylabel(w2.value)
    plt.title(w1.value + ' and ' + w2.value)
    #prevent data from displaying on each button click
    #plt.close(fig)
    return fig
#button aspects
def on_click(event):
    #call function to build plot
    #message.value = "on_click"
    fig = show_plot(w1, w2, color)
    message.value = "on_click"
    #save to png for display
    fig.savefig('plot.png')
    message.value = "after_save"
    #display image on screen
    display(Image('plot.png'))
    message.value = "after_Image"
button = pn.widgets.Button(name='Show Graph')
button.param.watch(on_click, 'clicks')
message = pn.widgets.TextInput(name='Message:', value='')
#add widgets created to next in column
column.append(w1)
column.append(w2)
column.append(color)
column.append(button)
column.append(message)
column
```
